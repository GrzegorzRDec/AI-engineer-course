**Numpy** opiera się na tabelach `arrays`. Powstały one do obliczeń matematycznych. Przypominają listy, ale wspomagają grupowe operacje matematyczne będące operacjami w `Numpy`.

```python
import numpy as np
```

```python
distance = [10, 15, 17, 26]
time = [.30, .47, .55, 1.20]
```

```python
np_distance = np.array(distance)
np_time = np.array(time)
speed = np_distance/np_time
```

```python
speed
```

`ndarray` na których opiera się Numpy cechuje się tym, że:
- jest kontenerem wartości
- jest homogeniczny tj. posiada tylko jeden typ danych w sobie
- można na nim operować normalnie, jak na liście tj. dodawać elementy, usumować, zmieniać...
- szybki i wydajny
- wspiera operacje matematyczne
- może być wielowymiarowy

```python
#first NumPy array
firstNpArray = np.array([1,2,3,4])

#printing
print(firstNpArray)
```

```python
#array with zeros
arrayWithZeros = np.zeros((3,3))
arrayWithZeros
```

```python
#array with ones
arrayWithOnes = np.ones((3,3))
arrayWithOnes
```

```python
#array with empty
arrayWithEmpty = np.empty((2,4))
arrayWithEmpty
```

```python
#array with arange method
npArange = np.arange(12)
print( npArange )
```

```python
#reshape:
npArange.reshape(3,4)
```

```python
#linspace for linearly spaced data elements
npLinear = np.linspace(1,6,5) #1 - wartość elementu pierwszego, 6 - wartość elementu ostatniego, 5 - ilość elementów
```

```python
print(npLinear)
```

```python
#One dimensional array
oneDArray = np.arange(15)
#reshape into 2D
twoDAttay = oneDArray.reshape(3,5)

print(twoDAttay)
```

```python
#three dimensional array
threeDArray = np.arange(27).reshape(3,3,3)
print(threeDArray)
```

**Operacje arytmetyczne na tablicach**

```python
#utworzenie tablic
pierwszaAr = np.array([1,11,111,2,22,222])
drugaAr = np.array([3,33,333,44,44,444])
```

```python
#dodawanie:
print( pierwszaAr + drugaAr )
```

```python
#odejmowanie:
print( drugaAr - pierwszaAr )
```

```python
#mnożenie dwóch wektorów:
print( pierwszaAr * drugaAr )
```

```python
#mnożenie przez skalar:
print( 5 * pierwszaAr )
```

```python
#podliczenie wektora:
print( drugaAr.sum() )
```

```python
#lub..... 
print( sum(drugaAr) )
```

**Operacje porównania na tablicach oraz indeksowanie wartościami logicznymi typu bool**

```python
#proste porównanie 
print(pierwszaAr > 20)
```

```python
#sprawdzenie, które wartości spełniają warunek:
pierwszaAr[ pierwszaAr > 20 ]
```

**Operacje logiczne na tablicach**

```python
#dla wartości różnych od zera jest True
np.logical_and(pierwszaAr, drugaAr)
```

# Funkcje matematyczne na tablicach

`Numpy` dostarcza szereg uniwersalnych funkcji matematycznych np.:
- `cos` kosinus dla każdego elementu tablicy
- `exp` eksponenta dla każdego elementu tablicy
- `floor` zwraca największą wartość całkowitą dla każdego elementu tabeli
- `sqrt` wiadomo - pierwiastek

```python
#floor
np.floor( [ 1.9, 2.4, -0.3, 5.1 ] )
```

**Operacje na rozmiarach tablicy**

```python
#stworzenie przykładowej tablicy:
przykladowa = np.array([[10,17,55,23,14,29],[92,14,54,34,23,33]])
print(przykladowa)
```

- `ravel`:

```python
#spłaszczanie danych:
print(przykladowa.ravel())
```

```python
#warto zwrócić uwagę na to, że nie wpływa na oryginalną tablicę:
print(przykladowa)
```

- `reshape`:

```python
#zwrócenie przekształconej tabeli
print(przykladowa.reshape(3,4))
```

```python
#bez wpływu na oryginał
print(przykladowa)
```

- `resize`:

```python
#zmiana oryginału
print(przykladowa.resize(3,4))
```

```python
print(przykladowa)
```

- `hsplit`:

```python
#podział w pionie
podprzyklad1, podprzyklad2 = np.hsplit(przykladowa,2)
```

```python
print(podprzyklad1)
```

```python
print(podprzyklad2)
```

- `hstack`:

```python
#połączenie:
powrot = np.hstack([podprzyklad1, podprzyklad2]) #<- tu może być też tupla, nie musi być lista
```

```python
print(powrot)
```

**Transpozycja**

```python
print(przykladowa)
```

```python
print(przykladowa.transpose())
```

**Inwersja**

```python
#przykladowa macierz kwadratowa
kwadrat = np.array([[10,20],[15,25]])
print(np.linalg.inv(kwadrat))
```

**Suma elementów po przekątnej**

```python
print(np.trace(kwadrat)) #w tym przypadku 10 + 25
```
