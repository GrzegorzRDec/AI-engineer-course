# Operacje na szeregach

**dodawanie wartosci po indexach**

```python
# dodawanie z indeksowaniem po kolei
s_pie = pd.Series([1,2,3,4],index=['a','b','c','d'])
s_dru = pd.Series([10,22,33,14],index=['a','b','c','d'])
```

```python
s_pie+s_dru
```

```python
#dodawanie z indeksowaniem na wspak
s_dru = pd.Series([10,22,33,14],index=['d','b','a','c'])
s_pie+s_dru
```

Jak widać wynik jest różny. Elementy zostały dodane według indeksów. 
Czyli np. 'a' (w s_pie = 1) do 'a' (w s_dru = 33) dają wynik 34.
<br />
<br />
W przypadku pomyłki z indeksami:

```python
s_dru = pd.Series([10,22,33,14],index=['d','b','f','g']) #<-- inne indeksy
s_pie+s_dru
```

Dla pojedynczych indeksów zwrócone zostaje `NaN`, czyli Not a Number.

# Operacje na danych

```python
import pandas as pd
#przykladowe dane:
dfMovieRating = pd.DataFrame({'movie 1':[5,4,3,3,2,1],'movie 2':[4,5,2,3,4,2]},index=['Tom','Jeff','Peter','Ram','Ted','Paul'])
dfMovieRating
```

```python
def movieGrade(rating):
    if rating == 5:
        return 'A'
    elif rating == 4:
        return 'B'
    elif rating == 3:
        return 'C'
    else:
        return 'F'
    
print(movieGrade(5))    
```

```python
#zastosowanie funkcji dla każdego elementu ramki
dfMovieRating.applymap(movieGrade)
```

# Standaryzacja danych

```python
#obliczenia odchylenia standardowego:
def standardizeTest(test):
    return (test-test.mean())/test.std()

standardizeTest(dfTestScores['Test1'])
```

```python
# przeprowadzenie obliczeń dla całej ramki:
def standaridizeTestScores(dataFrm):
    return dataFrm.apply(standardizeTest)

#standaryzowana ramka:
standaridizeTestScores(dfTestScores)
```

# Merge, Duplicate oraz contatenation

```python
import pandas as pd
```

```python
dfStudentMath = pd.DataFrame({'student':['Tom','Jack','Dan','Ram','Jeff','David'],
                              'ID':[10,56,31,85,9,22]})

dfStudentScience = pd.DataFrame({'student':['Tom','Ram', 'David'],
                                 'ID':[10,12,22]})
```

**Merge**

```python
#Po prostu połączenie dla pasujących 'student' jak i 'ID'
pd.merge(dfStudentMath,dfStudentScience)
```

```python
#Połączenie dla 'student', zwraca oba 'ID'
pd.merge(dfStudentMath, dfStudentScience, on='student')
```

```python
# Coś na zasadzie leftJoin na 'ID' oraz uzupełnienie wartości NaN poprzez 'X'
pd.merge( dfStudentMath, dfStudentScience, on='ID', how='left').fillna( 'X' )
```

```python
# Coś na zasadzie rightJoin na 'ID' oraz uzupełnienie wartości NaN poprzez 'X'
pd.merge( dfStudentMath,dfStudentScience, on='ID', how='right' ).fillna( 'X' )
```

**Concatenate** czyli powiązać

```python
#połączenie dwóch ramek
pd.concat( [dfStudentMath, dfStudentScience], ignore_index=True )
```

# SQL w Pandas


```python
import sqlite3
```

```python
#utworzenie tabeli SQL:
createTable = """
CREATE TABLE studentScore
( Id INTEGER, Name VARCHAR(20), Math REAL, Science REAL
);"""
```

```python
#wykonanie polecenia SQL:
executeSQL = sqlite3.connect(':memory:')
executeSQL.execute(createTable)
executeSQL.commit() # <-- tabela jest tworzona dopiero tutaj
```

```python
#przygotowanie zapytania:
sqlQuery = executeSQL.execute('select * from studentScore')
```

```python
resultset = sqlQuery.fetchall()
resultset
```

```python
#wprowadzenie danych do tabeli:
insertSQL = [(10,'Jack',85,92),
             (29,'Tom',73,89),
             (65,'Ram',65.5,77),
             (5,'Steve',55,91)]
```

```python
#wprowadzenie danych do tabeli:
insertStatement = "Insert into studentScore values(?,?,?,?)"
executeSQL.executemany( insertStatement, insertSQL )
executeSQL.commit()
```

```python
#przygotowanie zapytania:
sqlQuery = executeSQL.execute('select * from studentScore')
resultset = sqlQuery.fetchall()
resultset
```

```python
#zapis z SQL do ramki danych:
dfStudentSQL = pd.DataFrame( resultset, columns = next(zip( *sqlQuery.description )))
dfStudentSQL
```
