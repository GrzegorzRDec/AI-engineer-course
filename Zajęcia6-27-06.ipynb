# Scikit-learn





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn 
```

```python
dataset = pd.DataFrame({'Country':['France','Germany','Spain','France','Germany','Spain','France','Germany'],
                        'Age':[22,34,25,None,25,26,28,29],
                        'Salary':[72000,81000,10000,15000,None,17000,19000,22000],
                        'Purchased':['Yes','No','Yes','No','Yes','No','Yes','No']})
dataset
```

Independent variables: **Country**, **Ages**, **Salary**  *<- base on which we want to predict*

Dependent variables: **Purchased** *<- parameter to predict*



```python
X = dataset.iloc[:,:-1].values #everything except the last column with dependent variable
print(X)
print("\n"+str(type(X)))
```

```python
Xx = dataset.iloc[:,:-1]
Xx
```

```python
Y = dataset.iloc[:,-1].values
Y
```

# taking care of missing values:

methods inside sklearn: `Fit`  and  `fit_transform`

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer( missing_values = np.nan, strategy = 'mean' )
imputer.fit(X[:,1:3])
X #nic nie jest zmienione jeszcze
```

```python
X[:,1:3] = imputer.transform(X[:,1:3]) #dopiero ta metoda zamienia
X[:,1:3]
```

```python
X
```

```python
print(Xx['Age'].mean()) 
print(Xx['Salary'].mean())
```

**To samo można zrobić w jednej funkcji: fit_transform()**

```python
Xft = imputer.fit_transform(X[:,1:3])
Xft
```

# One hot encoding for indepedent variable

czyli 1 z n:
K1 | K2 | K3
 0 | 0  | 1
 0 | 1  | 0
 1 | 0  | 0

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer( transformers = [('encoder', OneHotEncoder(), [0] )], remainder = 'passthrough' )
X = np.array( ct.fit_transform(X) )
print(X)
```

Jak widać pierwsza kolumna została zamieniona na trzy, pozostałe zostały przepisane ('passthrough')

# label encoding in the dependent variable:

# Feature scaling

Two main methods:
Normalization: x_new = ( x_old - x_min ) / ( x_max - x_min )
Standarization: x_new = ( x_old - mean ) / std

**for 90% it's better to choose standarization!!!!** It is the advice, but it is like that.
And one method for all column

```python
#standarisation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform( X_train[ :, 3: ] ) # <-- omiting one hot coding columns
X_test[:,3:] = sc.transform( X_test[ :, 3: ] ) # <-- only transform (we are keeping the fitted values)

print(X_train)
print(X_test)
```
